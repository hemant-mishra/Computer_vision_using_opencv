```python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture(0) #reading from camera

while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if(ret==True):
        gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        cv.imshow('frame',gray)
        cv.waitKey(50)
        if(cv. getWindowProperty('frame',0)==-1): #if window closed
            break


# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
print("done")
```

```python
import cv2 as cv

cap = cv.VideoCapture('video_sample.avi')

# Define the codec and create VideoWriter object
fourcc = cv.VideoWriter_fourcc(*'DIVX')
out = cv.VideoWriter('output.avi',fourcc, 10, (1280,720))
#to have same output size as input, modify and use below code for mentioning width and height
'''
print(cap.get(3))
print(cap.get(4))

'''

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv.flip(frame,0)

        # write the flipped frame
        out.write(frame)
         
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv.destroyAllWindows()
print("done")
```
